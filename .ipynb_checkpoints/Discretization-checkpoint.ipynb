```python
# Creates a function that maps a number to a percentile,
# given a list of 'boundaries'.
# discretize
def discretize(sample, feature, percentile_ranks=[25,50,75]):
    feature_values = sample[feature]
    percentiles = [np.percentile(feature_values,x) for x in percentile_ranks]
    print(percentiles)
    discrete_output = []
    for x in list(feature_values):
        discrete_value = None
        
        for idx,percentile in enumerate(percentiles):
            if x <= percentile and discrete_value is None:
                discrete_value = idx
                break;
                
        if discrete_value is None:
            discrete_value = len(percentiles)
        
        discrete_output.append(discrete_value)
    
    return discrete_output

# netflow is the unencoded netflow feature vector
# mappings is an array of functions and/or dicts that map unencoded features to discrete values
# n_values is the number of output values for feature mappings, kept separate for code flexibility
def netflow_encoding(samples, features=['d_Prot','d_Packets'], verbose=False):
    code_list = []
    num_unique_values = {}
    
    for feature in features:
        if samples[feature].dtype.name == "Category":
            len(samples[feature].cat.categories)
        else:
            num_unique_values[feature] = samples[feature].nunique()
    
    counter = 0
    for _,item in samples.iterrows():
        counter = counter + 1
        if verbose:
            print ("Processing",counter,"out of", len(samples))
        code = 0
        space_size = np.prod([num_unique_values[feature] for feature in features])
        for feature in features:
            code = code + item[feature] * space_size/num_unique_values[feature]
            space_size = space_size/num_unique_values[feature]
        code_list.append(int(code))
        
    return code_list

#

def elbow(data, ns, title=""):
    data = np.array(data).reshape(-1,1)
    dist = DistanceMetric.get_metric('euclidean')

    errors = {}
    for n in ns:
        kmeans = KMeans(n_clusters=n).fit(data)
        centers = kmeans.cluster_centers_

        print('n={}'.format(n), end="\r", flush=True)

        errors[n] = sum(np.min(dist.pairwise(data, centers), axis=1))

    print()#newline

    plt.figure()
    plt.plot(errors.keys(), errors.values(), 'x-')
    plt.xticks(np.arange(min(ns), max(ns) + 1, 1.0))
    plt.ylabel('Summed distance')
    plt.xlabel('Nr of Clusters')
    plt.title(title)
    plt.show()
```

```python
# Create elbow plots for both features to see what the optimal number of clusters is
# df_conf[c], label_mapping[c] = pd.factorize(data[c])
df_conf = infected_flow_sample.sample(frac=0.4)
elbow(df_conf['Bytes'], range(1,11), "ELBOW plot of bytes")
elbow(df_conf['Packets'], range(1,11), "ELBOW plot of packets")
elbow(df_conf['Durat'], range(1,11), "ELBOW plot of duration")
# Discretize the infected host
```

```python
infected_host = infected_hosts[0]
infected_flow_sample = pd.DataFrame(df_clean.loc[(df_clean['Src_IP'] == infected_host)])
```

```python
# create a list of timed events
timed_events = []
prev_timestamp = None
for _,item in infected_flow_sample.iterrows():
    if prev_timestamp is None:
        time_delta = 0
    else:
        time_delta = int((item['Date_Time'] - prev_timestamp).total_seconds() * 1000) # save milliseconds
    # update prev_timestamp for next row
    prev_timestamp = item['Date_Time']
    
    timed_events.append((item['event_id'],time_delta))
```

```python
timed_events = {}
for host_ip in infected_hosts + legit_hosts:
    host_flows = pd.DataFrame(df_clean.loc[(df_clean['Src_IP'] == host_ip)])
    
    timed_events[host_ip] = []
    
    prev_timestamp = None
    for _,item in host_flows.iterrows():
        if prev_timestamp is None:
            time_delta = 0
        else:
            time_delta = int((item['Date_Time'] - prev_timestamp).total_seconds() * 1000) # save milliseconds
        # update prev_timestamp for next row
        prev_timestamp = item['Date_Time']

        timed_events[host_ip].append((item['event_id'],time_delta))
```
