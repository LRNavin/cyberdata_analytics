```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
# test_size: what proportion of original data is used for test set
```

```python
training_df = pd.read_csv('data/BATADAL_dataset03.csv', encoding="utf-8-sig",  delimiter=',', skipinitialspace=True, parse_dates=True, index_col='DATETIME', dayfirst=True)
testing_df = pd.read_csv('data/BATADAL_dataset04.csv', encoding="utf-8-sig",  delimiter=',', skipinitialspace=True, parse_dates=True, index_col='DATETIME', dayfirst=True)
testing_df.head()
```

```python
def intersection(lst1, lst2): 
    lst3 = [value for value in lst1 if value in lst2] 
    return lst3 

def diff(first, second):
        second = set(second)
        return [item for item in first if item not in second]
```

```python
# training_df['DATETIME']=pd.to_datetime(training_df.DATETIME,errors='coerce')
# testing_df['DATETIME']=pd.to_datetime(testing_df.DATETIME,errors='coerce')
training_df_1 = training_df.drop(columns=['ATT_FLAG'])
constants = [x for x in training_df_1 if np.std(training_df_1[x]) == 0]
all_headers = list(training_df_1)
features = diff(all_headers, constants)
features
```

```python
'''
Function that scales training and testing dataset into a standard scale form fitted on the training dataset
'''

def scale(df_train, df_test, features):
    scaled_features_df_train = df_train.copy()
    scaled_features_df_test = df_test.copy()
    
    feat_df_train = scaled_features_df_train[features]
    feat_df_test = scaled_features_df_test[features]
    scaler = StandardScaler().fit(feat_df_train.values)
    feat_df_train = scaler.transform(feat_df_train.values)
    feat_df_test = scaler.transform(feat_df_test.values)
    scaled_features_df_train[features] = feat_df_train
    scaled_features_df_test[features] = feat_df_test
    return scaled_features_df_train, scaled_features_df_test
```

```python
scaled_features_df_train, scaled_features_df_test = scale(training_df,testing_df,features)


scaled_features_clean_df_train = scaled_features_df_train.drop(columns=['ATT_FLAG'])
scaled_features_clean_df_test = scaled_features_df_test.drop(columns=['ATT_FLAG'])
pca = PCA()
pca.fit(scaled_features_clean_df_train)
```

```python
scaled_features_clean_df_train = pca.transform(scaled_features_clean_df_train)
scaled_features_clean_df_test = pca.transform(scaled_features_clean_df_test)

print (np.sort(pca.explained_variance_)[::-1])
```

```python
cummulative_variance=cummulative_variance=pca.explained_variance_ratio_.cumsum()
print(cummulative_variance)
```

```python
x_axis = np.arange(1, scaled_features_clean_df_train.shape[1]+1, 1)
plt.xlabel('Principal Components')
plt.ylabel('Cummulative Variance Captured')
plt.plot(x_axis, cummulative_variance)
```

```python
#Keep test labels
labels = np.array(testing_df['ATT_FLAG'].tolist())
i=0
for el in labels:
    if el==1:
        i+=1
print(i)
```

```python
del testing_df['ATT_FLAG']
testing_df
```

```python
scaler = StandardScaler()
scaled_test_set = scaler.fit_transform(testing_df)

#Normalize test set
pca_test = pca.transform(scaled_test_set)
```

```python
#Find residuals
eigenvectors = pca.components_
P = np.transpose(eigenvectors[:-5])
P_T = np.transpose(P)
C = np.dot(P, P_T)
I = np.identity(testing_df.shape[1])

# y_residual corresponds to the projection of test data on anomalous subspace
y_residual = np.zeros(scaled_test_set.shape)

# Find projection of test data on anomalous subspace
for i in range(scaled_test_set.shape[0]):
    y = np.transpose(scaled_test_set[i])
    y_residual[i] = np.dot(I - C, y)
```

```python
#Calculate SPE for each residual
DateOfAnomalyPCA=[]
spe = np.zeros((scaled_test_set.shape[0]))
attack = np.ones((scaled_test_set.shape[0]))
threshold = 40
for i in range(scaled_test_set.shape[0]):
    spe[i] = np.square(np.sum(np.subtract(y_residual[i], scaled_test_set[i])))     
    if(spe[i] > threshold):
        attack[i] = 1
        date = testing_df.index[i].strftime("%Y-%m-%d %H:%M:%S")
        
        if date not in DateOfAnomalyPCA:
            DateOfAnomalyPCA.append(i)
        
    else:
        attack[i]=-999
```

```python

thres=[]
for i in range(spe.shape[0]):
    thres.append(threshold)
```

```python

#Residuals Plot
testing_df = testing_df.assign(ResidualVector=spe)
testing_df = testing_df.assign(Threshold=thres)
testing_df['ResidualVector'].plot(figsize=(15,5),label="ResidualVector")
testing_df['Threshold'].plot(figsize=(15,5), label="Threshold")
plt.legend(bbox_to_anchor=(0.7, 1), loc=2)
plt.xlabel("DATETIME")
plt.ylabel("Residual Vector")
plt.show()
```

```python

#Calculate Metrics
from sklearn.metrics import confusion_matrix
tn,fp,fn,tp=confusion_matrix(labels, attack).ravel()
```

```python

print ("TP:", tp)
print ("FP:", fp)
print ("FN:", fn)
print ("TN:", tn)
```

```python

Accuracy=(tp+tn)/(tp+tn+fp+fn)*100
Precision=tp / (tp + fp)*100
Recall = tp / (tp + fn)*100
F_score = 2*tp /(2*tp + fp + fn)*100
print("Accuracy:","%.2f" % Accuracy)
print("Precision:","%.2f" % Precision)
print("Recall:", "%.2f" %Recall)
print("F_score:","%.2f" % F_score)
```
